
> The main objective of this work is to understan how Functional models work.

```python
# import libraries
import torch
import torch.nn as nn
import pandas as pd
import matplotlib.pyplot as plt
import torch.nn.functional as F
import warnings
warnings.filterwarnings("ignore")
```

```python
# read a csv
df = pd.read_csv('prostate-cancer-1.csv')
```

```python
# print top 5 rows
df.head()
```

```python
# define the train set
x_train = torch.tensor(df.drop('Y', axis=1)[:71].values).float() # train
y_train = torch.tensor(df['Y'][:71].values).float() # label

# Define the test set
x_test = torch.tensor(df.drop('Y', axis=1)[71:].values).float() # 
y_test = torch.tensor(df['Y'][71:].values).float()
```

```python
class fn_model(torch.nn.Module):
    def __init__(self):
        super(fn_model, self).__init__()
        self.linear1 = torch.nn.Linear(x_train.shape[1], 50)
        self.linear2 = torch.nn.Linear(50, 1)
    def forward(self, x):
        relu1 = F.relu(self.linear1(x))
        out_put = F.relu(self.linear2(relu1))
        return out_put
```

```python
model = fn_model()
```

```python
# dfine the loss function
loss_function = torch.nn.MSELoss()
```

```python
# define the optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
# train the model for n_iter number of iteration
n_iter = 2000
for step in range(n_iter):
    y_pred = model(x_train).squeeze()
    loss = loss_function(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if step%500 == 0: # print every 500 steps
        print(f"Step {step} : loss {round(loss.item(), 4)}")
```

```python
# Test
pred = model(x_test)
predicted = pd.DataFrame({'Ground_truth': [i.item() for i in y_test],\
                          "Prediction": [i.item() for i in pred]})
```
